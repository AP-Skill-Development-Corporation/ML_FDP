```python
Ridge
- add penalty term
- complex
- model performance will be low
Lasso
- removes the unwanted features
- loss of information
ElasticNet
- combination of lasso and ridge
- costfunction+(lambda(slope)^2)+lambda(|slope|)
- works well with the large data set
```

```python
import pandas as pd
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df["Target"]=boston.target
boston_df
```

```python
boston_df.isna().sum()
```

```python
x=boston_df.iloc[:,0:13:1]
y=boston_df["Target"]
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_scale=scale.fit_transform(x)
x_scale
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.2,random_state=100)
from sklearn.linear_model import ElasticNet
model=ElasticNet()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
model.coef_
```

#### KNN(K Nearest Neighbors)

- choose the value of k
- find out the distances
- compare the k nearest neighbors
- predict the ouput which got majority vote


```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Classification/shirt.csv")
```

```python
df
```

```python
x=df[["Height","Weight"]]
y=df["Size"]
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=4)  # by calculating the sqrt(n),n=4
knn.fit(x,y)
```

```python
y_pred=knn.predict(x)
```

```python
knn.predict([[161,61]])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

```python
fruits_df
```

```python
fruits_df.shape
```

```python
x=fruits_df.iloc[:,2:]
y=fruits_df["fruit_label"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)  # by calculating the sqrt(n),k=7
knn.fit(x,y)
```

```python
y_pred=knn.predict(x_test)
```

```python
knn.predict([[3,192,8.4,7.3,0.55]])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
